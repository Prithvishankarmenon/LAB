```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#load a image
image =cv2.imread('/home/y4ada2/Downloads/rose.jpeg',0)
#Apply histogram Equilisation

equilised_image=cv2.equalizeHist(image)

#display both

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('original image ')


plt.subplot(1,2,2)
plt.imshow(equilised_image,cmap='gray')
plt.title('equilised image')


plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.hist(image)
plt.title('original image Histogram')

plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.hist(equilised_image)
plt.title('equilised image Histogram')

plt.show()
```

```python
#morphological operation (erosion)
kernel=np.ones((5,5),np.uint8)
#input image
image=cv2.imread('/home/y4ada2/Downloads/Alpha.jpg',0)
#use erode
eroded_image=cv2.erode(image,kernel,iterations=1)
#display
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('original image ')

plt.subplot(1,2,2)
plt.imshow(eroded_image,cmap='gray')
plt.title('eroded image')

plt.show()
```

```python
#morphological operation (dilation)
kernel=np.ones((5,5),np.uint8)
#choose image
image=cv2.imread('/home/y4ada2/Downloads/Alpha.jpg',0)
#Dilate image
dilated_image=cv2.dilate(image,kernel,iterations=1)
#display image
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title('original image ')

plt.subplot(1,2,2)
plt.imshow(dilated_image,cmap='gray')
plt.title('dilated image')

plt.show()
```
