## Part 1 

```python
import seaborn as sns
from scipy.stats import binom
import matplotlib.pyplot as plt

data = binom.rvs(n=10,p=0.5,size=1010)

sns.set_style('whitegrid')
ax = sns.distplot(data, kde=True, color='red' )

ax.set(xlabel='Binomial',ylabel='Frequency')

plt.show()
```

## Part 2

```python
import seaborn as sns
from scipy.stats import poisson
import matplotlib.pyplot as plt

data = poisson.rvs(mu=17,size=100)

sns.set_style('whitegrid')
ax = sns.distplot(data, kde=True, color='red' )

ax.set(xlabel='Poisson',ylabel='Frequency')

plt.show()
```

# Exp 10 

```python
import scipy.stats as stats
import numpy as np
data_group1 = np.array([10,12,0,15,15,20,15,18,12,13,11,19,17,15,13,14])
data_group2 = np.array([25,45,11,23,12,1,2,45,12,13,13,19,17,15,88,14])
print(np.var(data_group1),np.var(data_group2))
stats.ttest_ind(a = data_group1,b = data_group2,equal_var = True)
```

# Exp 11

```python
import scipy.stats as stats
import numpy as np

observed_data = [11,34,33,5,23,7]
expected_value = [25,25,25,31,2,5]

chi_square_test_statistic,p_value = stats.chisquare(observed_data, expected_value)

print("Chi square test Statistic ",chi_square_test_statistic)
print("P_value ",p_value)
```
